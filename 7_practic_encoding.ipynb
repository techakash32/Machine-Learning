```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head(2)
```

```python
df=df.dropna()
```

```python
df=df.drop(columns=['day'])
```

```python
## APPROACH-1. : LabelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['sex']=lb.fit_transform(df['sex'])
df['smoker']=lb.fit_transform(df['smoker'])
df['time']=lb.fit_transform(df['time'])
```

```python
df.head(2)
```

```python
## APPROACH-2: OrdinalEncoding
```

```python
df=sns.load_dataset('tips')
```

```python
df.head(2)
```

```python
df['time'].value_counts()
```

```python
df=df.drop(columns=['total_bill','tip','day','size'])
```

```python
df=df.dropna()
```

```python
df_new=df.copy()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['Female','Male'],
                               ['No','Yes'],
                               ['Dinner','Lunch']])
```

```python
oe_df=oe.fit_transform(df)
```

```python
df_oe=pd.DataFrame(oe_df,columns=df_new.columns)
```

```python
df_oe
```

```python
## APPROACH 3: ONEHOTENCODER
```

```python
df=sns.load_dataset('tips')
```

```python
new=df.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
df
```

```python
oh=ohe.fit_transform(df[['sex','smoker','day','time']])
```

```python
df=pd.DataFrame(oh,columns=ohe.get_feature_names_out(['sex','smoker','day','time']))
```
